```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
import plotly.express as px 
```

```python
hr_df=pd.read_csv('/content/HRDataset_v14.csv')
hr_df.head(5)
```

```python
hr_df.info()
hr_df.columns
```

## Descriptive stats about the whole HR data frame 


```python
hr_df.describe()
```

## Mean- max, median of salary, absences and various surveys 

```python
hr_df[['Salary', 'EngagementSurvey','EmpSatisfaction','Absences']].describe()
```

```python
## From above we can conclude that Salary has mean=69020 and the max value and min value is as followed= 25,0000 and 45046
## Absences we have max of 20 and min is 1 while the average of absenses in the company is 10
```

## **Salary distribution** 

```python
fig=plt.figure(figsize=(8,5))
salary_distribution=sns.displot(hr_df.Salary,kde='True')
salary_distribution.figure.suptitle('Distribution of Salary')
```

# Absence distribution 

```python
fig=plt.figure(figsize=(8,5))
absence_distribution=sns.displot(hr_df.Absences,kde=True)
absence_distribution.figure.suptitle('Distribution of Absences')
```

# Distribution of Employee satisfaction

```python

fig=plt.figure(figsize=(8,5))
empsat_distribution=sns.displot(hr_df.EmpSatisfaction,kde='True')
empsat_distribution.figure.suptitle('Distribution of Employee satisfaction')
```

# Distribution of special projects assigned 

```python
fig=plt.figure(figsize=(8,5))
splproj_distribution=sns.displot(hr_df.SpecialProjectsCount)
splproj_distribution.figure.suptitle('Distribution of Special projects assigned ')
```

# Relation between Employee's and Salary

```python
eSalary=sns.relplot(x='EmpID',y='Salary',data=hr_df)
eSalary.figure.suptitle('Relation of Employees and Salary')
```

```python
# conclusion:- the max salary is 250000 and average is around 50,000
```

```python
hlSal=sns.relplot(x='EmpID',y='Salary',hue='HispanicLatino',data=hr_df)
hlSal.figure.suptitle('Ethnicity of Employess whether or not Hispanic or Latina')
```

```python
## Conclusion:- The higesht paid employee is Hispanic/Latina and Most employees are not Hispanic/Latino
```

# Salary and Department Comparison 

```python
sns.barplot(x='Salary',y='Position',data=hr_df)
```

```python
## Conclusion :- CIO is highest paid job and Production Techncian is low salary job
```

# Status of Citizenship of employees

```python
sns.relplot(x='EmpID',y='Salary',hue='CitizenDesc',data=hr_df)
```

# Distribution of Employees citizenship 

```python
sns.countplot(y='CitizenDesc',data=hr_df)
```

```python
fig = plt.figure(figsize=(20,20))
sns.relplot(x='EmpID',y='Salary',hue='CitizenDesc',col='MaritalDesc',data=hr_df)
plt.title('Employees vs salary and analysing marital status as well as citizenship status')
plt.show()
```

```python
## Conclusion:- Employees who are Divorced,Widowed or Seprated are US citizen/ Eligible US citizen
```
